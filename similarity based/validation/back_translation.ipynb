```python
from googletrans import Translator
translator = Translator()
```

```python
from tqdm import tqdm
i =0
result = [0] * len(df)
#df = pd.read_csv('/content/drive/MyDrive/파일위치/pan_qna_trans.csv')
#result = list(df['trans_question'])
for txt in tqdm(df['question'][i:]):
    try:
        result[i] = translator.translate(translator.translate(txt, src='ko', dest='en').text, src='en', dest='ko').text
    except:
        pass
    i += 1
    if i % 1000 == 0:
        df['trans_question'] = result
        df.to_csv('/content/drive/MyDrive/파일위치/pan_qna_trans.csv')
```
